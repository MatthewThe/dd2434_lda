```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filename = "output_Lambda.tsv"
mat = np.loadtxt(filename)
```

```python
plt.figure(figsize=(20,10))
plt.imshow(mat, aspect="auto")
plt.title("Topics vs Vocabulary", fontsize=20)
plt.ylabel("Topics", fontsize=20)
plt.xlabel("Vocabulary", fontsize=20)
plt.savefig('topic_vocab.png')
```

```python
filename = "output_gamma.tsv"
gamma_matrix = np.loadtxt(filename)
print(gamma_matrix.shape)

plt.figure(figsize=(20,10))
plt.imshow(gamma_matrix, aspect="auto")
plt.title("Gamma(Documents vs Topics)", fontsize=20)
plt.ylabel("Documents", fontsize=20)
plt.xlabel("Topics", fontsize=20)
plt.savefig('gamma.png')
```

```python
filename = "output_theta_original.tsv"
theta_matrix = np.loadtxt(filename)
print(theta_matrix.shape)

plt.figure(figsize=(20,10))
plt.imshow(theta_matrix, aspect="auto")
plt.title("Original Theta (Documents vs Topics)", fontsize=20)
plt.ylabel("Documents", fontsize=20)
plt.xlabel("Topics", fontsize=20)
plt.savefig('theta.png')
```

```python
plt.figure(figsize=(20,10))
plt.imshow(gamma_matrix, aspect="auto")
plt.imshow(theta_matrix, aspect="auto")
plt.title("Gamma vs Theta", fontsize=20)
plt.ylabel("Documents", fontsize=20)
plt.xlabel("Topics", fontsize=20)
plt.savefig('gamma_theta.png')
```

```python
filename = "output_z_original.tsv"
z_matrix = np.loadtxt(filename)

K = 3
D,N = z_matrix.shape

new_z = np.zeros((D,K))

for d in range(D):
    for k in range(K):
        temp = np.where(z_matrix[d,:]==k)[0]
        count = len(temp)
        new_z[d,k] = count
        
new_z = new_z.astype(int)


filename = "output_gamma.tsv"
gamma_matrix = np.loadtxt(filename)

plt.figure(figsize=(20,10))

plt.subplot(121)
plt.imshow(new_z, aspect="auto")
plt.title('Z', fontsize=20)
plt.xlabel('Topics', fontsize=20)
plt.ylabel('Documents', fontsize=20)

plt.subplot(122)
plt.imshow(gamma_matrix, aspect="auto")
plt.xlabel('Topics', fontsize=20)
plt.title('Gamma', fontsize=20)

plt.show()
```

```python
D,K = new_z.shape

ordered_z = np.zeros((D,K))
ordered_z[:,0] = new_z[:,0]
ordered_z[:,1] = new_z[:,2]
ordered_z[:,2] = new_z[:,1]

plt.figure(figsize=(20,10))

plt.subplot(121)
plt.imshow(ordered_z, aspect="auto")
plt.title('Z', fontsize=20)
plt.xlabel('Topics', fontsize=20)
plt.ylabel('Documents', fontsize=20)

plt.subplot(122)
plt.imshow(gamma_matrix, aspect="auto")
plt.xlabel('Topics', fontsize=20)
plt.title('Gamma', fontsize=20)

plt.show()
```

```python
filename = "output_theta_original.tsv"
theta_matrix = np.loadtxt(filename)

filename = "output_gamma.tsv"
gamma_matrix = np.loadtxt(filename)

plt.figure(figsize=(20,10))

plt.subplot(121)
plt.imshow(theta_matrix, aspect="auto")
plt.title('Theta', fontsize=20)
plt.xlabel('Topics', fontsize=20)
plt.ylabel('Documents', fontsize=20)

plt.subplot(122)
plt.imshow(gamma_matrix, aspect="auto")
plt.xlabel('Topics', fontsize=20)
plt.title('Gamma', fontsize=20)

plt.show()
```
